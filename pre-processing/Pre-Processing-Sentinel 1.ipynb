```python
# Set target folder and extract metadata - Change accordingly
product_path = str(r'D:\\MICROS\\Sentinel_1\\')

input_S1_files =[]
for item in os.listdir(product_path):
    if item.endswith('zip'):
        input_S1_files.append(item)
 
   
name, process_level, sensing_date, relevant_orbit, height, width, band_names = ([] for i in range(7))

for i in input_S1_files:
    process_level.append(i.split("_")[2])
    sensing_date.append(i.split("_")[4][0:-7])
    relevant_orbit.append(i.split("_")[-2])
    # Read with snappy
    s1_read = snappy.ProductIO.readProduct(product_path + i)
    name.append(s1_read.getName())
    height.append(s1_read.getSceneRasterHeight())
    width.append(s1_read.getSceneRasterWidth())
    band_names.append(s1_read.getBandNames())
    
    
    
#     width = s1_read.getSceneRasterWidth()
#     height = s1_read.getSceneRasterHeight()
#     name = s1_read.getName()
#     description = s1_read.getDescription()
#     band_names = s1_read.getBandNames()
#     starttime = s1_read.getStartTime()
#     endtime = s1_read.getEndTime()
    
df_s1_read = pd.DataFrame({'Name': name, 'Process_Level': process_level, 'Sensing_Date': sensing_date, 'Relevant_Orbit': relevant_orbit, 'Height': height, 'Width': width, 'Band_Names': band_names})
display(df_s1_read)
```

```python
outpath_name = str('D:\\MICROS\\Sentinel_1\\Subsets\\Limassol\\')

# Create the output folder if not already exist
if not(os.path.exists(outpath_name) and os.path.isdir(outpath_name)):
    os.makedirs(outpath_name)


# Iterate through the product list  
for i in input_S1_files[49:]:
    
    s1_read = snappy.ProductIO.readProduct(product_path + i)    # read the product
    name = s1_read.getName()
    output = str(outpath_name) + str(name)
    print(f'Reading product {name}\n')
    
    subset = snap.subset(s1_read) # Subset Operator - snappy 


    cal = snap.calibration(subset) # Calibration 
    orbit = snap.applyorbitfile(cal) # Apply Orbit File
    terrain = snap.terraincorrection(orbit) # Terrain Correction

    snappy.ProductIO.writeProduct(terrain, output+".tif", 'GeoTIFF')  
    print(f'Saving the new product on disk...')
    time.sleep(4)
    s1_read.closeIO()
    del s1_read, subset,cal,orbit, terrain
    gc.collect()
    
```
